# Entity Recognition Models
This notebook is about testing some models on entity recognition, which is the task of determining whether a certain term in text refers to some entity.

```python
%config IPCompleter.greedy=True
import pandas as pd
```

```python
# data generated from https://github.com/Ryan-Amaral/wikisim/blob/master/wikification/ml-model.ipynb
df1 = pd.read_csv('er-10000-cls-neg.txt', header=None)
df2 = pd.read_csv('er-10000-cls-pos.txt', header=None)
df = pd.concat([df1,df2])
df.columns = ['posBef', 'posOn', 'posAft','mentionProb',
              'stanford','cap','match','space','ascii','isMention']
df = df.drop(columns=['stanford']) # get rid of stanford scores so just our model
df = df.sample(frac=1, random_state=42).reset_index(drop=True) # https://stackoverflow.com/questions/29576430/shuffle-dataframe-rows
df.head()
```

##### Each record is a token (part of text) described by the following:


posBef: Part of Speech tag before the token.

posOn: Part of Speech tag of the token.

posAft: Part of Speech tag after the token.

mentionProb: Probability that the token is a mention/link on Wikipedia.

cap: Whether the token is capitalized.

match: Whether there is an exact title match in wikipedia.

space: Whether there is a space in the token.

ascii: Whether the token is ascii only.

isMention: Whether the token is a mention.

```python
# dummify the pos ones
df = pd.get_dummies(df, columns=['posBef', 'posOn', 'posAft'])
df.head()
```

```python
df.describe() # gotta fix that mention prob max
```

```python
df.mentionProb = df.mentionProb.clip(upper=1)
df.describe()
```

```python
# do 70-30 train test split
rows = len(df.index)
y = df.loc[:,'isMention'].values
X = df.drop('isMention',1).loc[:,:].values
print(X.shape,y.shape)
from sklearn.model_selection import train_test_split
XTrain, XTest, yTrain, yTest = train_test_split( X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

print('adaboost:')
abc = GridSearchCV(AdaBoostClassifier(), 
    {'n_estimators': [100,200,300]}).fit(XTrain,yTrain)

print('bagging:')
bgc = GridSearchCV(BaggingClassifier(), 
    {'n_estimators': [100,200,300]}).fit(XTrain,yTrain)

print('gradient boosting:')
gbc = GridSearchCV(GradientBoostingClassifier(), 
    {'n_estimators': [100,200,300]}).fit(XTrain,yTrain)

print('random forest:')
rfc = GridSearchCV(RandomForestClassifier(), 
    {'n_estimators': [100,200,300]}).fit(XTrain,yTrain)
```

```python
print('AdaBoost', abc.best_params_)
print('Bagging', bgc.best_params_)
print('Gradient Boosting', gbc.best_params_)
print('Random Forest', rfc.best_params_)
```

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score

print('adaboost report:')
print(classification_report(yTest, abc.predict(XTest)))
cvAbc = cross_val_score(abc, XTest, yTest, cv=5)
print(cvAbc)

print('bagging report:')
print(classification_report(yTest, bgc.predict(XTest)))
cvBgc = cross_val_score(bgc, XTest, yTest, cv=5)
print(cvBgc)

print('gradient boosting report:')
print(classification_report(yTest, gbc.predict(XTest)))
cvGbc = cross_val_score(gbc, XTest, yTest, cv=5)
print(cvGbc)

print('random forest report:')
print(classification_report(yTest, rfc.predict(XTest)))
cvRfc = cross_val_score(rfc, XTest, yTest, cv=5)
print(cvRfc)
```

```python
import matplotlib.pyplot as plt
ax = plt.boxplot([cvAbc,cvBgc,cvGbc,cvRfc], 
           labels=['AdaBoost','Bagging','Gradient Boost','Random Forest'])
plt.title('ER Classification Performance')
plt.show()
```

```python
# data generated from https://github.com/Ryan-Amaral/wikisim/blob/master/wikification/ml-model.ipynb
df1 = pd.read_csv('er-10000-cls-neg.txt', header=None)
df2 = pd.read_csv('er-10000-cls-pos.txt', header=None)
df = pd.concat([df1,df2])
df.columns = ['posBef', 'posOn', 'posAft','mentionProb',
              'stanford','cap','match','space','ascii','isMention']
df = df.drop(columns=['posBef','posAft','stanford']) # get rid of stanford scores so just our model
df = df.sample(frac=1, random_state=42).reset_index(drop=True) # https://stackoverflow.com/questions/29576430/shuffle-dataframe-rows
df.head()
# try it all again, to see how powerful without pos before and after
```

```python
# dummify the pos ones
df = pd.get_dummies(df, columns=['posOn'])
df.head()
```

```python
df.mentionProb = df.mentionProb.clip(upper=1)
df.describe()
```

```python
# do 70-30 train test split
rows = len(df.index)
y = df.loc[:,'isMention'].values
X = df.drop('isMention',1).loc[:,:].values
print(X.shape,y.shape)
from sklearn.model_selection import train_test_split
XTrain, XTest, yTrain, yTest = train_test_split( X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier

print('adaboost:')
abc1 = AdaBoostClassifier(n_estimators=200)
abc1.fit(XTrain,yTrain)

print('bagging:')
bgc1 = BaggingClassifier(n_estimators=300)
bgc1.fit(XTrain,yTrain)

print('gradient boosting:')
gbc1 = GradientBoostingClassifier(n_estimators=300)
gbc1.fit(XTrain,yTrain)

print('random forest:')
rfc1 = RandomForestClassifier(n_estimators=200)
rfc1.fit(XTrain,yTrain)
```

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score

print('adaboost report:')
print(classification_report(yTest, abc1.predict(XTest)))
cvAbc = cross_val_score(abc1, XTest, yTest, cv=5)
print(cvAbc)

print('bagging report:')
print(classification_report(yTest, bgc1.predict(XTest)))
cvBgc = cross_val_score(bgc1, XTest, yTest, cv=5)
print(cvBgc)

print('gradient boosting report:')
print(classification_report(yTest, gbc1.predict(XTest)))
cvGbc = cross_val_score(gbc1, XTest, yTest, cv=5)
print(cvGbc)

print('random forest report:')
print(classification_report(yTest, rfc1.predict(XTest)))
cvRfc = cross_val_score(rfc1, XTest, yTest, cv=5)
print(cvRfc)
```

```python
import matplotlib.pyplot as plt
ax = plt.boxplot([cvAbc,cvBgc,cvGbc,cvRfc], 
           labels=['AdaBoost','Bagging','Gradient Boost','Random Forest'])
plt.title('ER Classification Performance')
plt.show()
```
